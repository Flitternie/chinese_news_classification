```python
import os
import re
import json
import pandas as pd

doc_topics = ['会议', '会见', '调研', '活动', '批示']
```

```python
def read_file(filepath):
    with open(filepath, 'r', encoding='gb18030') as f:
        content = f.read()
        content = re.sub('[\n\t\s]*', '', content)
        if '#' not in content:
            print('{}'.format(filepath))
            return None
    return content.split('#', 1)
```

```python
doc_topic_df = pd.DataFrame(columns=['topic_id', 'content', 'file'])      
paragraph_topic_df = pd.DataFrame(columns=['topic_id', 'content', 'file'])  

for doc_topic in doc_topics:
    topic_folders = os.listdir("./new_data/{}".format(doc_topic))
    topic_folders = [folder for folder in topic_folders if folder.isdigit()]
    topic_folders.sort()
    for folder in topic_folders:
        files = os.listdir("./new_data/{}/{}/".format(doc_topic, folder))
        content = read_file("./new_data/{}/{}/type.txt".format(doc_topic, folder))
        if content:
            doc_topic_df = doc_topic_df.append({'topic_id': int(content[0].split('-')[0])-1, 'content': content[1], 'file': "{}.{}".format(doc_topic, folder)}, ignore_index=True)
        files = [file for file in files if re.match(r'p[0-9]*\.txt$', file)]
        for file in files:
            content = read_file("./new_data/{}/{}/{}".format(doc_topic, folder, file))
            if content:
                paragraph_topic_df = paragraph_topic_df.append({'topic_id': int(content[0].split('-')[0])-1, 'content': content[1], 'file': "{}.{}.{}".format(doc_topic, folder, file)}, ignore_index=True)
```

```python
doc_topic_df.to_json(open('val_doc_topic.json', 'w', encoding='utf-8'), force_ascii=False, orient='records')
doc_topic_df.groupby('topic_id').count()
```

```python
paragraph_topic_df.to_json(open('val_paragraph_topic.json', 'w', encoding='utf-8'), force_ascii=False, orient='records')
paragraph_topic_df.groupby('topic_id').count()
```

```python
def read_test_file(filepath):
    with open(filepath, 'r', encoding='gb18030') as f:
        content = f.read()
        content = re.sub('[\n\t\s]*', '', content)
    return content
```

```python
test_doc_topic_df = pd.DataFrame(columns=['content', 'file'])      
test_paragraph_topic_df = pd.DataFrame(columns=['content', 'file'])  

topic_folders = os.listdir("./test_data/")
topic_folders = [folder for folder in topic_folders if folder.isdigit()]
topic_folders.sort()
for folder in topic_folders:
    files = os.listdir("./test_data/{}/".format(folder))
    content = read_test_file("./test_data/{}/type.txt".format(folder))
    if content:
        test_doc_topic_df = test_doc_topic_df.append({'content': content, 'file': "{}".format(folder)}, ignore_index=True)
    files = [file for file in files if re.match(r'p[0-9]*\.txt$', file)]
    for file in files:
        content = read_test_file("./test_data/{}/{}".format(folder, file))
        if content:
            test_paragraph_topic_df = test_paragraph_topic_df.append({'content': content, 'file': "{}.{}".format(folder, file)}, ignore_index=True)

test_doc_topic_df.to_json(open('test_doc_topic.json', 'w', encoding='utf-8'), force_ascii=False, orient='records')
test_paragraph_topic_df.to_json(open('test_paragraph_topic.json', 'w', encoding='utf-8'), force_ascii=False, orient='records')
```
